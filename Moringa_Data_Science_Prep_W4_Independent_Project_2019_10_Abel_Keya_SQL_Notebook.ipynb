## 1.1 Importing Data from CSV files

```
# Loading our table from the respective CSV files
# We will export the Households dataset that we have cleaned using python
#
with open('autolib.csv','r') as f:
    autolib = pd.read_csv(f, index_col=0, encoding='utf-8')
%sql DROP TABLE if EXISTS autolib;
%sql PERSIST autolib;
%sql SELECT * FROM autolib LIMIT 1;
```

```
%%sql
SELECT*  FROM autolib
WHERE city = "Paris"
ORDER BY hour DESC;
```

```
# Let's preview the first 10 rows of our data
%%sql 
SELECT* FROM autolib
ORDER BY 'hour' LIMIT 10
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
%%sql
SELECT hour,city
FROM(SELECT*  FROM autolib
      WHERE city = "Paris"
      )
GROUP  BY hour
ORDER  BY COUNT(*) DESC LIMIT 1;  
```

```
#What is the most popular hour for returning cars?
%%sql
SELECT  hour
FROM autolib     
GROUP  BY hour
ORDER  BY COUNT(*) ASC LIMIT 1; 
```

```
#What station is the most popular?Overall?
%%sql
SELECT Kind,Geopoint,ID
FROM(SELECT*  FROM autolib
      WHERE city = "Paris"
      )
GROUP  BY ID
ORDER  BY COUNT(*) ASC LIMIT 1; 
```

```
#What station is the most popular?At the most popular picking hour?
%%sql
SELECT Kind,Geopoint,ID
FROM
(SELECT hour,ID,Kind,Geopoint
FROM(SELECT*  FROM autolib
      WHERE city = "Paris"
      )
GROUP  BY hour
ORDER  BY COUNT(*) DESC)
GROUP  BY ID
ORDER  BY COUNT(*) DESC LIMIT 1;
```

```
#What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
#Overall?
%%sql
SELECT Postalcode
FROM
(SELECT *
FROM(SELECT*  FROM autolib
      WHERE city = "Paris"
      )
GROUP  BY hour
ORDER  BY COUNT(*) DESC)
GROUP  BY Cars
ORDER  BY COUNT(*) DESC LIMIT 1; 
```

```
#What postal code is the most popular for picking up Bluecars? Does the most popular station belong to that postal code?
#At the most popular picking hour?
%%sql
SELECT Postalcode
FROM
(SELECT hour,ID,Postalcode
FROM(SELECT*  FROM autolib
      WHERE city = "Paris"
      )
GROUP  BY hour
ORDER  BY COUNT(*) DESC)
GROUP  BY Postalcode
ORDER  BY COUNT(*) DESC LIMIT 1;
```
