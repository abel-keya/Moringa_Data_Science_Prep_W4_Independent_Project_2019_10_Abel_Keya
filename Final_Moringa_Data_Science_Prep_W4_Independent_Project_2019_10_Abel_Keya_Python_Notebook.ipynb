```

# Procedure 3: Irrelevant Data Observation
# Data Cleaning Action: Dropping Status attribute
# Explanation: I won't need it during Analysis
autolib.drop('Status',axis=1,inplace=True)
```

```
#add datetime colunm in the standard format
autolib['datetime'] = autolib['day'].map(str) + '/' + autolib['month'].map(str) + '/' + autolib['year'].map(str) + ' ' + autolib['hour'].map(str) + ':' + autolib['minute'].map(str)
#add datetime colunm in the standard format
autolib['datetime'] = pd.to_datetime(autolib['datetime'])
autolib
```
